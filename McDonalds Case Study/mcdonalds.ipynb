```python
def calulate_mean(data):
    # selecte the relevant columns
    selected_columns = data.iloc[:,0:11]
    # Create a new dataframe to store the modified values
    new_data = pd.DataFrame(index=selected_columns.index,
                            columns=selected_columns.columns)
    # Loop through each column and calculate the mean
    for col in selected_columns.columns:
        for i in range(len(selected_columns)):
            if selected_columns.at[i,col]=='Yes':
                new_data.at[i,col] = 1
            elif selected_columns.at[i,col]=="No":
                new_data.at[i,col] = 0
            else:
                new_data.at[i,col]=selected_columns.at[i, col]
    # Calculate mean for each column
    mean_values = {}
    for col in new_data.columns:
        column_sum=0
        for value in new_data[col]:
            column_sum += value
            mean_values[col]=round(column_sum/len(new_data),2)
        
    return pd.Series(mean_values)
mean_result = calulate_mean(data)
print(mean_result)
   
```

```python
# Replace "Yes" with 1 and "No" with 0 using loops
MD = data.iloc[:, 0:11].copy()  # Copy the first 11 columns to avoid modifying the original data
for column in MD.columns:
    MD[column] = MD[column].apply(lambda x: 1 if x == "Yes" else (0 if x == "No" else x))

# Calculate mean of each column and round to 2 decimal places
mean = {}
for column in MD.columns:
    mean[column] = round(MD[column].mean(), 2)

mean  
```

```python
# Perform PCA
pca = PCA()
MD_pca = pca.fit_transform(MD)
MD_p = pca.fit(MD)

# Calculate standard deviation and proportion of variance
SD = np.sqrt(pca.explained_variance_)
PV = pca.explained_variance_ratio_

# Create an index for principal components
index = []
for i in range(len(SD)):
    index.append(f"PC{i+1}")

# Create a DataFrame to summarize PCA results
sum_df = pd.DataFrame({
    "Standard deviation": SD,
    "Proportion of Variance": PV,
    "Cumulative Proportion": PV.cumsum()
}, index=index)

sum_df
```

```python
np.random.seed(1234)
# No of repetitaions for KMeans and range of cluster numbers
nrep = 10
num_segments = range(1, 9)
# INitialize lists and dictionaries to sotre results
within_cluster_distances = []
MD_km28 = {}
# Loop through each value of k (number of clusters)
for k in num_segments:
    # Perform KMeans clustering
    kmeans = KMeans(n_clusters=k, n_init=nrep, random_state=1234)
    kmeans.fit(MD)
     # Store the within-cluster sum of squared distances (inertia)
    within_cluster_distances.append((kmeans.inertia_))
    # Store the fitted KMeans model in a dictionary
    MD_km28[str(k)] = kmeans
# Plot the results
plt.bar(num_segments, within_cluster_distances)
plt.xlabel("Number of segments")
plt.ylabel("Sum of within-cluster distances")
plt.title("Segmentation Results")
plt.show()
```

```python
np.random.seed(1234)
nboot=100 # No of bootstrap samples
nrep=10 # no of repetitions for KMeans
# Create bootstrap samples
bootsrap_samples=[]
for i in range(nboot):
    # Create a bootstrap sample
    bootstrap_sample = resample(MD.values,random_state=1234)
    bootsrap_samples.append(bootstrap_sample)
# index
adjusted_rand_index=[]
num_segments=range(2,9)
for k in num_segments:
    stability_scores=[]
    # Initialize the list to store the adjusted rand index for each bootstrap sample
    for bootstrap_sample in bootsrap_samples:
        kmeans=KMeans(n_clusters=k,n_init=nrep, random_state=1234)
        kmeans.fit(bootstrap_sample)
        # Preditct cluster labels for bootstrap sample and original data
        cluster_labels =kmeans.predict(bootstrap_sample)
        true_lables=kmeans.predict(MD.values)
        # Calulate stability score and add to the list
        stability_score = adjusted_rand_score(true_lables, cluster_labels)
        stability_scores.append(stability_score)
    # Append all stability scores for the current k values
    adjusted_rand_index.append(stability_scores)
# Converting the list of Ari values to a numpy array and transporse it
adjusted_rand_index = np.array(adjusted_rand_index).T
# Box plot 
plt.figure(figsize=(10,8))
plt.boxplot(adjusted_rand_index, labels=num_segments, whis=10)
plt.xlabel("Number of Clusters (Segments)")
plt.ylabel("Adjusted Rand Index (Stability Score)")
plt.title("Bootstrap Clustering Stability (Flexclust)")
plt.show()
```

```python
from scipy.stats import entropy

np.random.seed(1234)
k_values = range(2, 9)
MD_m28 = []

for k in k_values:
    model = KMeans(n_clusters=k, random_state=1234)
    model.fit(MD.values)
    iter_val = model.n_iter_
    converged = True
    k_val = k
    k0_val = k
    log_likelihood = -model.inertia_
    n_samples, _ = MD.shape
    aic = -2 * log_likelihood + 2 * k
    bic = -2 * log_likelihood + np.log(n_samples) * k
    labels = model.labels_
    counts = np.bincount(labels)
    probs = counts / float(counts.sum())
    class_entropy = entropy(probs)
    icl = bic - class_entropy
    
    MD_m28.append((iter_val, converged, k_val, k0_val, log_likelihood, aic, bic, icl))
MD_m28 = pd.DataFrame(MD_m28, columns=['iter', 'converged', 'k', 'k0', 'logLik', 'AIC', 'BIC', 'ICL'])

print(MD_m28)
```

```python
num_segments = MD_m28["k"]
AIC_values = MD_m28["AIC"]
BIC_values = MD_m28["BIC"]
ICL_values = MD_m28["ICL"]

plt.plot(num_segments, AIC_values, marker='o', label='AIC')
plt.plot(num_segments, BIC_values, marker='o', label='BIC')
plt.plot(num_segments, ICL_values, marker='o', label='ICL')

plt.xlabel('Number of Segments')
plt.ylabel('Value of Information Criteria')
plt.title('Information Criteria (AIC, BIC, ICL)')
plt.legend()
plt.grid(True)

plt.show()
```
