```python
average_price = data.groupby('Location')['Price'].transform(lambda x: x.mean())
data['Average Price By City'] = average_price
data.head()


```

```python
# average price in each city
average_price = data.groupby('Location')['Price'].mean()
Price=average_price.values
# Create Data frame
City_avgPrice_df = pd.DataFrame(zip(l1,Price),
                                columns=['City','Average Price in City(Lakhs)'])
City_avgPrice_df
```

```python
# Sort values for better visualization
City_avgPrice_df = City_avgPrice_df.sort_values(by="Average Price in City(Lakhs)", ascending=True)

# Set color palette using Seaborn
colors = sns.color_palette("coolwarm", len(City_avgPrice_df))

# Plot
plt.figure(figsize=(12, 8))
bars = plt.barh(City_avgPrice_df['City'], 
                City_avgPrice_df['Average Price in City(Lakhs)'],
                color=colors)

# Add data labels
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, 
             f"{bar.get_width():.2f}", va='center', ha='right', fontsize=11)

# Styling
plt.xlabel("Average Price in City (Lakhs)", fontsize=13)
plt.ylabel("City", fontsize=13)
plt.title("Average Car Price in Each City", fontsize=15, fontweight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.6)

# Show plot
plt.show()
```

```python
# city vs fule type vs Price
# Ensure the data is sorted for better visualization
City_fuel_df = data.groupby(['Location', 'Fuel_Type'])['Price'].mean().reset_index()

# Set figure size
plt.figure(figsize=(14, 7))

# Create a grouped bar plot
sns.barplot(x='Location', y='Price', hue='Fuel_Type', data=City_fuel_df, palette='viridis')

# Rotate x-axis labels for readability
plt.xticks(rotation=45, ha='right')

# Labels and title
plt.xlabel("City", fontsize=13)
plt.ylabel("Average Price (Lakhs)", fontsize=13)
plt.title("Average Price of Cars by Fuel Type in Each City", fontsize=15, fontweight='bold')

# Show grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.stripplot(x='Location', y='Price', hue='Fuel_Type', data=data, palette='coolwarm', jitter=True, alpha=0.7)

plt.xticks(rotation=45, ha='right')
plt.xlabel("City", fontsize=13)
plt.ylabel("Price (Lakhs)", fontsize=13)
plt.title("Distribution of Car Prices by Fuel Type in Each City", fontsize=15, fontweight='bold')

plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.boxplot(x='Location', y='Price', hue='Fuel_Type', data=data, palette='coolwarm')

plt.xticks(rotation=45, ha='right')
plt.xlabel("City", fontsize=13)
plt.ylabel("Price (Lakhs)", fontsize=13)
plt.title("Price Distribution by Fuel Type in Each City", fontsize=15, fontweight='bold')

plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
City_fuel_pivot = data.pivot_table(index='Location', columns='Fuel_Type', values='Price', aggfunc='mean')

plt.figure(figsize=(18, 10))
sns.heatmap(City_fuel_pivot, cmap="coolwarm", annot=True, fmt=".1f", linewidths=0.5,annot_kws={"fontsize": 13, "fontweight": "bold"})

plt.xlabel("Fuel Type", fontsize=14,fontweight='bold')
plt.ylabel("City", fontsize=14,fontweight='bold')
plt.title("Heatmap of Average Car Prices by Fuel Type in Each City", fontsize=15, fontweight='bold')

plt.show()
```

Key Observations

Diesel Cars Have Higher Average Prices

Diesel vehicles generally have higher average prices across most cities, with Coimbatore (18.8 lakhs) and Bangalore (15.7 lakhs) showing the highest prices.
The price range for diesel cars varies significantly, indicating regional demand differences.

Electric Cars Show Considerable Variation

The price of electric cars is quite variable, with Chennai (13.0 lakhs) and Mumbai (12.8 lakhs) having higher averages.
Some cities, such as Coimbatore, show a lack of electric vehicle price data, which may indicate lower adoption.

CNG and LPG Cars Are More Affordable

CNG vehicles have the lowest average prices, typically ranging from 2.3 lakhs (Hyderabad) to 9.5 lakhs (Kolkata and Bangalore).
LPG vehicles are the least expensive, with prices between 1.4 lakhs (Mumbai) and 6.0 lakhs (Jaipur).
This suggests that LPG and CNG options remain budget-friendly choices in most cities.

Petrol Car Prices Are Moderate

Petrol vehicle prices mostly range between 4.2 lakhs (Jaipur) and 8.2 lakhs (Coimbatore).
Prices are relatively stable across cities, showing consistency in market pricing.

City-Wise Highlights

Coimbatore has the highest average diesel car price (18.8 lakhs).
Bangalore has relatively higher prices across fuel types, particularly for diesel (15.7 lakhs) and CNG (9.5 lakhs).
Kolkata has a high average price for CNG vehicles (9.5 lakhs), similar to Bangalore.
Mumbai and Chennai show significant pricing for electric vehicles (12.8 lakhs and 13.0 lakhs, respectively).
Hyderabad and Pune have lower-priced CNG and LPG cars compared to other cities.

Conclusions

Diesel cars tend to be the most expensive across cities, especially in Coimbatore and Bangalore.
Electric cars have varying price points, indicating different levels of market penetration.
CNG and LPG cars are the most budget-friendly, making them an affordable choice in multiple cities.
Petrol cars maintain a stable price range, showing balanced demand and availability.

```python
# Plot histogram
plt.figure(figsize=(12, 6))
sns.histplot(data, x='Price', hue='Fuel_Type', kde=True, bins=20, palette='coolwarm', alpha=0.7,)
for fuel in data['Fuel_Type'].unique():
    sns.kdeplot(data[data['Fuel_Type']==fuel]['Price'],
                label=fuel,linewidth=2.5)
# Bold Labels
plt.xlabel("Price (Lakhs)", fontsize=13, fontweight='bold')
plt.ylabel("Count", fontsize=13, fontweight='bold')
plt.title("Price Distribution by Fuel Type", fontsize=15, fontweight='bold')

# Show plot
plt.show()
```

Observations

Price Distribution:

The majority of vehicles are priced within the lower range (0-25 lakhs), with a steep decline in the number of vehicles as the price increases.
The distribution follows a right-skewed pattern, indicating that most vehicles are affordable, while only a few are priced at the higher end.

Fuel Type Comparison:

Petrol and Diesel vehicles dominate the dataset, contributing significantly to the price distribution.
CNG vehicles show a smaller distribution, indicating limited availability or adoption.
LPG and Electric vehicles appear to have a minimal presence in the dataset, possibly due to lower market penetration.
Electric vehicles exhibit a wider price range, with a few models extending beyond 100 lakhs, highlighting the presence of premium-priced EVs.

Density Trends:

The kernel density estimation (KDE) lines suggest that petrol and diesel cars have similar price distributions, with peaks around lower price ranges.
The price distribution for electric vehicles is spread out more, reflecting diverse pricing across budget and luxury segments.
